# Part 1 - Violin Plot - Distribution of settlement/closure amounts based on Disposition
Remember, do not include data where the claim was denied

```python
#Acquire general idea of field data types by outputting dtypes of df14.
df14.dtypes
```

```python
#Create new dataframe with df14 data to avoid manipulating original dataframe
df14_2 = pd.read_csv("assets/claims-2014.csv")
```

```python
#remove all rows in data frame where closing amount has no value, since those are irrelevant.
df14_2['Close Amount'] = df14_2['Close Amount'].str.replace("-","")
df14_2['Close Amount'].replace('', np.nan, inplace=True)
df14_2['Close Amount'] = df14_2['Close Amount'].astype('float64')
df14_2.dropna(subset=['Close Amount'], inplace=True)
```

```python
#remove all rows in data frame where disposition is 'Deny', as requested by management
#now only fields with 'Approve in Full' and 'Settle' as values will remain
df14_2.drop(df14_2.loc[df14_2['Disposition']=='Deny'].index, inplace=True)
```

```python
#Now create the violin plots using seaborn facetgrid functionality and output plot
#Min closing amount = 1.25 - found with .min() function
#Max closing amount = 25483.44 - found with .max() function
#Mean closing amount is 226.28 - found with .mean() function
fig, ax = plt.subplots(figsize =(9, 7)) 
sns.violinplot(ax = ax, x = df14_2['Disposition'],  y = df14_2['Close Amount'] ).set_title('Close Amount by Final Outcome') 
```

Based on the violin plots shown above, one can see that fully approved compensation/settlement amounts - in general - don't really tend to venture into outrageous territory. The compensation value with the highest probability of occurrence for fully approved verdicts, based on the width of the leftmost boxplot, seems to be visually close to the minimum Closing Amount of 1.25 dollars. The compensation value with the highest probability of occurrence for settlements, based on the width of the rightmost boxplot, seems to be visually close to about 1000 dollars, though it is difficult to tell. The most salient feature of these plots, however, are the outlier values. The largest court approved compensation appears visually close to about $8000, while the largest settlement is even more extravagent, close to 26,000. From a purely monetary standpoint, it appears as if settling outside of court should be avoided, if only to avoid the potentiality of more debilitating costs.

# Part 2 - Bar plot - Total Damages per Month based on Incident Date 

```python
#create new dataframe with df14 data to avoid manipulating original dataframe
#remove all rows where close amount as no value
df14_3 = pd.read_csv("assets/claims-2014.csv")
df14_3['Close Amount'] = df14_3['Close Amount'].str.replace("-","")
df14_3['Close Amount'].replace('', np.nan, inplace=True)
df14_3['Close Amount'] = df14_3['Close Amount'].astype('float64')
df14_3.dropna(subset=['Close Amount'], inplace=True)
```

```python
#Convert Date Received column to datetime and then extract only year out of date. Get rid of years that don't == 2014
#After deletion, extract the month from the full date and use for bar plot
df14_3['Incident Date'] = df14_3['Incident Date'].astype('datetime64[ns]')
df14_3['Incident Date'] = df14_3['Incident Date'].astype('datetime64[ns]')
df14_3['Year Received'] = df14_3['Incident Date'].dt.year
df14_3 = df14_3[df14_3['Year Received'] == 2014]
df14_3['Month Received'] = df14_3['Incident Date'].dt.month
df14_3.head()

```

```python
#Now plot the data
sns.set(style="whitegrid")
barplot = sns.barplot(x="Month Received", y="Close Amount", data=df14_3, ci=None).set_title('Close Amount by Month - 2014')
```

The barplot shown above depicts the total number of incidents that have occured in order of month(Jan-Dec) for the year 2014. It is easy to discern that the month that incurred the greatest number of incidents is March, followed by August. It's possible to surmise that this is on account of the fact that their is a greater number of fliers in these two months of the year than others, though the month of March does not coincide with major holiday travel in the US. It will be necessary to analyze further historical data in order to ascertain whether or not this is a trend or an isolated incident.

# Part 3 - Boxplots - Top N Airports with Highest Total Settlement Amounts

```python
#First, validate the airports with the top 5 highest settlement costs in 2015
df15_2 = pd.read_csv("assets/claims-2015.csv")
df15_2['Close Amount'] = df15_2['Close Amount'].str.replace("-","")
df15_2['Close Amount'].replace('', np.nan, inplace=True)
df15_2['Close Amount'] = df15_2['Close Amount'].astype('float64')
df15_2.dropna(subset=['Close Amount'], inplace=True)
df15_2.drop(df15_2.loc[df15_2['Disposition']=='Deny'].index, inplace=True)
df15_2.drop(df15_2.loc[df15_2['Disposition']=='Approve in Full'].index, inplace=True)
df15_2['Incident D'] = df15_2['Incident D'].astype('datetime64[ns]')
df15_2['Incident D'] = df15_2['Incident D'].astype('datetime64[ns]')
df15_2['Year Received'] = df15_2['Incident D'].dt.year
df15_2 = df15_2[df15_2['Year Received'] == 2015]
df15_2 = df15_2.sort_values(by='Close Amount', ascending=False)
df15_2 = df15_2[:5]
df15_2.head(10)


```

```python
#Now, create another dataframe and only include those airports that were in the top 5 in terms of settlement costs
df15_3 = pd.read_csv("assets/claims-2015.csv")
df15_3['Settlement Amount'] = df15_3['Close Amount']
df15_3['Settlement Amount'] = df15_3['Settlement Amount'].str.replace("-","")
df15_3['Settlement Amount'].replace('', np.nan, inplace=True)
df15_3['Settlement Amount'] = df15_3['Settlement Amount'].astype('float64')
df15_3.dropna(subset=['Settlement Amount'], inplace=True)
df15_3.drop(df15_3.loc[df15_3['Disposition']=='Deny'].index, inplace=True)
df15_3.drop(df15_3.loc[df15_3['Disposition']=='Approve in Full'].index, inplace=True)
df15_3['Incident D'] = df15_3['Incident D'].astype('datetime64[ns]')
df15_3['Incident D'] = df15_3['Incident D'].astype('datetime64[ns]')
df15_3['Year Received'] = df15_3['Incident D'].dt.year
df15_3 = df15_3[df15_3['Year Received'] == 2015]
df15_3 = df15_3[df15_3['Airport Name'].isin(['Dallas Love Field', 'Nome', 'Metropolitan Oakland International', 'New Orleans International','Miami International Airport'])]

```

```python
boxplot = sns.boxplot(x="Airport Code", y="Settlement Amount", data=df15_3).set_title('Settlement Amount by Airport Code')
```

The above chart displays the spread of settlement amounts across the airports with the top five settlement costs in the year 2015(from left to right: Dallas Love Field	, Miami International Airport, New Orleans International, Metropolitan Oakland International, and Nome). Because only one observance existed for Alaska's Nome airport in 2015, a distribution spread could not be generated, though the settlement amount is still plotted. Of all airports, Dallas Love Field received the incident that resolved with the highest settlement, at nearly 2500 dollars, indicated by the outlier tick mark above the airport's box plot, followed by Nome, at 2362.25 dollars. It is worth noting that all of the cases that were resolved for an amount over $1000 were are all outlier values, which suggests that, in general, most incidices at even these airports fall within a predictable range - though analyzing historical data would be beneficial in order to determine the exact frequency of these isolated, outlier incidents.

# Part 4 - Your Choice Visualization - Comparing total number of claims per month, 2014 and 2015 

```python
df14 = pd.read_csv("assets/claims-2014.csv")
df15 = pd.read_csv("assets/claims-2015.csv")
```

```python
#Output head of df14 to view field names
df14.head(1)
```

```python
#Output head of df15 to view field names, and change field names to match df14 exactly to make concatenation easier
df15.rename(columns={"Incident D": "Incident Date"}, inplace = True)
df15.head(1)
```

```python
#remove all rows where close amount as no value
df14['Close Amount'] = df14['Close Amount'].str.replace("-","")
df14['Close Amount'].replace('', np.nan, inplace=True)
df14['Close Amount'] = df14['Close Amount'].astype('float64')
df14.dropna(subset=['Close Amount'], inplace=True)


df15['Close Amount'] = df15['Close Amount'].str.replace("-","")
df15['Close Amount'].replace('', np.nan, inplace=True)
df15['Close Amount'] = df15['Close Amount'].astype('float64')
df15.dropna(subset=['Close Amount'], inplace=True)
```

```python
#Convert Date Received column to datetime and then extract only year out of date. Get rid of years that don't == 2014 & 2015
#After deletion, extract the month from the full date to use for bar plot
#Create count of incidents to per month column using mapping dictionary for each dataframe
#As a final step, combine both dataframes using pd.concat
df14['Incident Date'] = df14['Incident Date'].astype('datetime64[ns]')
df14['Incident Date'] = df14['Incident Date'].astype('datetime64[ns]')
df14['Year Received'] = df14['Incident Date'].dt.year
df14 = df14[df14['Year Received'] == 2014]
df14['Month Received'] = df14['Incident Date'].dt.month
counts_1 = df14['Month Received'].value_counts().to_dict()
df14['Count of incidents']= df14['Month Received'].map(counts_1)



df15['Incident Date'] = df15['Incident Date'].astype('datetime64[ns]')
df15['Incident Date'] = df15['Incident Date'].astype('datetime64[ns]')
df15['Year Received'] = df15['Incident Date'].dt.year
df15 = df15[df15['Year Received'] == 2015]
df15['Month Received'] = df15['Incident Date'].dt.month
counts_2 = df15['Month Received'].value_counts().to_dict()
df15['Count of incidents']= df15['Month Received'].map(counts_2)



frames = [df14, df15]

df_concat = pd.concat(frames)






```

```python
#Now, plot a side by side bar chart comparing incidents per month for 2015 and 2015, using seaborn
fourteen_vs_fifteen = sns.barplot(x="Month Received", y="Count of incidents", hue="Year Received", data=df_concat).set_title('2014 vs. 2015: # of incidents per month')
```

# Justification

The double bar chart shown above displays the total number of incidents recorded per month(regardless of disposition) for 2014 and 2015, and plots them side-by-side. I believe that doing so facilitates much easier comparison than visually comparing two individual bar plots. Furthermore, since obtaining value counts was the main objective of this particlar analysis, a bar chart was a better designed medium to do so. Though a histogram is somewhat similar in function, since each group in this analysis is defined by a categorical variable(month) as opposed to a quantitative variable, it would not have been a feasible representation in this particular case.


At a glance, it is striking to see how many more incidents generally occured each month in 2015 compared to 2014. Only in April were the number of incidents higher in 2014 than 2015, albeit by a somewhat negligible amount. The months where the greatest number of incidents were observed are March and July, for 2014 and 2015 respectively, whereas the month where the least number of incidents was observed was December, for both 2014 and 2015.
